```python
# 1.  Establish the environment and set up the ranges for the data capture.

# Dependencies
# Make sure environment variable are setup
#  1. config import api_key
#  2. citipy import citipy
#  3. pip install citipy

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
#import time
#import json

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# 2.  Generate random coordinates using numpy to use for cities list
rand_coordinates = [( np.random.uniform(-90,90 ), np.random.uniform(-180,180) ) for k in range(1400)]
```

```python
#  3.  Define the cities dictionary = []
for coordinate_pair in rand_coordinates:
    lat, long = coordinate_pair
    cities.append(citipy.nearest_city(lat, long))
```

```python
#  4.  Define the cities_list = []
for city in cities:
    country_code = city.country_code
    name = city.city_name
    country = city.country_code
    cities_list.append(name)
```

```python
#  5.  Define the actual_cities_list = []
for city in cities_list:
    if city not in actual_cities_list:
        actual_cities_list.append(city)
```

```python
#  6.  Access the Open Weather Map org and collect url data via the OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

# Get current weather

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#  7. Create the lists to hold reponse info  (note: attempting to add country need to clean out)
city_name = []
country_name = []
lat = []
temperature = []
cloudiness = []
windspeed = []
humidity = []
count=1
for city in actual_cities_list:
    print(f"Processing Record number {count} | {city} | {country}")
    print(query_url + str(city) + str(country))
    count+=1
    
print("-------------------------------------")
print("Cities Data Retrieval Complete")
print("-------------------------------------")
```

```python
#   9. Loop through the list of cities and perform a request for data on each
for n in actual_cities_list:
    response = requests.get(query_url + str(n)).json()
    try:
        lat.append(response['coord']['lat'])
    except:
        continue
    city_name.append(response['name'])
    temperature.append(response['main']['temp'])
    windspeed.append(response['wind']['speed'])
    cloudiness.append(response['clouds']['all'])
    humidity.append(response['main']['humidity'])
```

```python
#  10.  Create the dataframe from parsing through the actual cities list and print the information
data_df=pd.DataFrame({"City":city_name,
                      "Latitude":lat,
                      "Temperature":temperature,
                      "Cloudiness":cloudiness,
                      "Windspeed":windspeed,
                      "Humidity":humidity,
                      })
data_df
```

```python
#  11.  Export the data from the dataframe to a file
exported_data=data_df.to_csv('Random_Weather_Data.csv')
```

```python
#  12.  Plot the Cities Latitudes vs Temperatures on a simple scatter plot
x_axis=data_df["Latitude"]
y_axis=data_df["Temperature"]
plt.scatter(x_axis,y_axis)
plt.xlabel("Latitude")
plt.ylabel("Temperature(F)")
plt.title("Latitude vs Temperature")
plt.savefig('Latitude_vs_Temp')
plt.show()
```
